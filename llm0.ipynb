```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

hi_text = "سلام. ما داریم راجب ترنسفومر ها صحبت میکنیم"
chinese_text = "生活就像一盒巧克力。"

model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")

# translate Persian to Eng
tokenizer.src_lang = "fa"
encoded_fa = tokenizer(hi_text, return_tensors="pt")
generated_tokens = model.generate(**encoded_fa, forced_bos_token_id=tokenizer.get_lang_id("en"))
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
# => "La vie est comme une boîte de chocolat."

```

```python

# translate Chinese to English
tokenizer.src_lang = "zh"
encoded_zh = tokenizer(chinese_text, return_tensors="pt")
generated_tokens = model.generate(**encoded_zh, forced_bos_token_id=tokenizer.get_lang_id("fa"))
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
# => "Life is like a box of chocolate."
```

```python
model_name = "deepset/roberta-base-squad2"
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer


model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")

def translate(text):
  tokenizer.src_lang = "fa"
  encoded_fa = tokenizer(text, return_tensors="pt")
  generated_tokens = model.generate(**encoded_fa, forced_bos_token_id=tokenizer.get_lang_id("en"))
  return ' '.join(tokenizer.batch_decode(generated_tokens, skip_special_tokens=True))
# => "La vie est comme une boîte de chocolat."
def translate_fa(text):
  tokenizer.src_lang = "en"
  encoded_fa = tokenizer(text, return_tensors="pt")
  generated_tokens = model.generate(**encoded_fa, forced_bos_token_id=tokenizer.get_lang_id("fa"))
  return ' '.join(tokenizer.batch_decode(generated_tokens, skip_special_tokens=True))

QA_input = {
    'question': 'تاریخ تولد ایلان ماسک',
    'context': 'ایلان ریو ماسک FRS (/ˈiːlɒn/؛ متولد 28 ژوئن 1971) یک تاجر و سرمایه‌گذار است که به خاطر نقش‌های کلیدی‌اش در شرکت فضایی SpaceX و شرکت خودروسازی تسلا، شرکت شناخته می‌شود. دیگر مشارکت‌ها شامل مالکیت X Corp.، شرکتی است که پلتفرم رسانه اجتماعی X (که قبلاً توییتر نامیده می شد) را اداره می کند و نقش خود را در تأسیس The Boring Company، xAI، Neuralink و OpenAI دارد. او یکی از ثروتمندترین افراد جهان است. تا اوت 2024، فوربس دارایی خالص او را 247 میلیارد دلار تخمین زد.[4]'
}
for k in QA_input :
  QA_input[k] = translate(QA_input[k])
# a) Get predictions
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)

res = nlp(QA_input)
res
```

```python
questions = ['اسم دقیق فرد معرفی شده چیست',
             'شغل این فرد چیست',
             'تاریخ تولد این فرد چه زمانی است','شرکت های این فرد چیست',
             'چقدر پول داره','نام پدر این فرد چیست']
ans = []
for q in questions:
  QA_input['question'] = translate(q)
  res = nlp(QA_input)
  ans.append(translate_fa(res['answer']))
```
